# Some Cycling Statistics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from celluloid import Camera
```

Let's import some data

```python
stats = pd.read_excel('bike_data.xlsx')
stats
```

```python
fig, ax = plt.subplots(1, 2, figsize=(25, 10), dpi=150)
cam = Camera(fig)

for m in range(len(stats['Date'])):

    ax[0].plot(stats['Date'][:m],
               stats['Distance'][:m],
               '^r',
               linestyle='solid')  # Distance in Km
    ax[0].set_ylabel('Distance (Km)')
    ax[0].set_ylim(5, 40)
    ax[0].set_title('Distance per day')

    ax[1].plot(stats['Date'][:m],
               stats['Distance'][:m] * 60 / stats['Time'][:m],
               '^g',
               linestyle='solid')  # speed
    ax[1].set_ylabel('Speed (Km/h)')
    ax[1].set_ylim(15, 25)
    ax[1].set_title('Speed per day')
    plt.suptitle(
        'Accumulated Distance = %3.2f Km \n Accumulated Time = %s:%s \n Number of days = %s'
        % (stats['Distance'].sum(), stats['Time'].sum() // 60,
           stats['Time'].sum() % 60, len(stats['Date'])))
    cam.snap()
cam.animate(blit=False, interval=500, repeat=True).save('anim.mp4')
```
